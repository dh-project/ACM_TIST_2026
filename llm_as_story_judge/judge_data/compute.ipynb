```python
import pandas as pd



# id valutati dagli umani
ids_historical = set(panel.loc[panel["judge"]=="humanHistorical", "id"])
ids_dramatic   = set(panel.loc[panel["judge"]=="humanDrama", "id"])

# ---- numeriche ----
historical_cols = [
    "historical_aderenzaEventoOriginale",
    "historical_plausibilita__EventiStorici",
    "historical_plausibilita__LuoghiAmbientazioni",
    "historical_plausibilita__PersonaggiStorici",
]
dramatic_cols = [
    "dramatic_potenzialeStorytelling",
    "dramatic_realizzazioneElementiDinamici",
    "dramatic_situazioneDrammatica",
]

panel_hist = panel[panel["id"].isin(ids_historical)]
panel_dram = panel[panel["id"].isin(ids_dramatic)]

# media + std per judge e generator
hist_stats = panel_hist.groupby(["generator","judge"])[historical_cols].agg(["mean","std"])
dram_stats = panel_dram.groupby(["generator","judge"])[dramatic_cols].agg(["mean","std"])

# ---- turning points: conteggi per judge e generator ----
tp = [
    ("climax", "dramatic_turningPoints__climaxPresente", "dramatic_turningPoints__climaxOrdinato"),
    ("eventoScatenante", "dramatic_turningPoints__eventoScatenantePresente", "dramatic_turningPoints__eventoScatenanteOrdinato"),
    ("risoluzione", "dramatic_turningPoints__risoluzionePresente", "dramatic_turningPoints__risoluzioneOrdinato"),
]

rows = []
for name, pres_col, ord_col in tp:
    for (gen, judge), df in panel_dram.groupby(["generator","judge"]):
        pres = df[pres_col].dropna().astype(float)
        pres_ord = ((df[pres_col] == True) & (df[ord_col] == True)).astype(float)

        rows.append({
            "generator": gen,
            "judge": judge,
            "dimension": name,
            "presente_count": int(pres.sum()),
            "presente_std": float(pres.std()) if len(pres) > 1 else 0.0,
            "presente_e_ordinato_count": int(pres_ord.sum()),
            "presente_e_ordinato_std": float(pres_ord.std()) if len(pres_ord) > 1 else 0.0,
        })

tp_counts_by_gen = pd.DataFrame(rows)
pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.4f}".format)


hist_stats
```
